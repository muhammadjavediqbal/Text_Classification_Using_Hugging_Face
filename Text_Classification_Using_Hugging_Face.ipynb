```python
from transformers import pipeline
```

```python
nlp = pipeline("text-classification")
```

```python
input_text = "Happy"
output = nlp(input_text)
```

```python
label = output[0]['label']
score = output[0]['score']
```

```python
print("Text:", input_text)
print("Sentiment:", label)
print("Confidence Score:", score)
```
